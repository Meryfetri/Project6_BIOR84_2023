```python
# bring in the packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import xlwings as xw

# interactivity (only in jupyter)
%matplotlib notebook

#load in the data
df = pd.read_csv("Sofia1.csv")

#axes definitions
x= df['x_cm']
y= df['y_cm']
z= df['z_cm']

#create the figure
fig = plt.figure(figsize = (6,6))
ax = Axes3D(fig)
fig.add_axes(ax)

# creat plot
ax.scatter(x,y,z, s=20, edgecolors='black')

# set axes names
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Create an empty workbook & rename sheet
#wb = xw.Book()
#sht=wb.sheets[0]
#sht.name="Python Plots"

#Insert plot into Excel
#sht.pictures.add(fig,
                #name = "Test 1",
                #update=True,
                #left=sht.range("A4").left,
                #top=sht.range("A4").top,
                #height=200,
                #width=300,)
```
